```python
df = df.drop_duplicates()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler=StandardScaler()
df[['Time', 'Amount']] = scaler.fit_transform(df[['Time', 'Amount']])
```

```python
feature= df.drop(columns=['Class'])
target = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    feature, target, test_size=0.2, random_state=42, stratify=target
)
```

```python
X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)
```

```python
import joblib
```

```python
joblib.dump(scaler,"scaler.pkl")
```

```python
import os
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv").squeeze()
y_test = pd.read_csv("y_test.csv").squeeze()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
X_normal = X_train[y_train == 0]
```

```python
X_normal.shape
```

```python
model = IsolationForest(
    n_estimators=400,
    contamination=0.008,
    max_samples=0.8,
    max_features=1.0,
    random_state=42)
model.fit(X_normal)
```

```python
joblib.dump(model, "model_forest.pkl")
```

```python
pred = model.predict(X_test)
pred_labels = np.where(pred == -1, 1, 0)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, pred_labels))
print("\nClassification Report:\n", classification_report(y_test, pred_labels, digits=4))
```

```python
auc_score = roc_auc_score(y_test, pred_labels)
print(f"ROC-AUC Score: {auc_score:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
```

```python
import matplotlib.pyplot as plt
```

```python

fpr, tpr, _ = roc_curve(y_test, pred_labels)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1], [0,1], 'r--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
